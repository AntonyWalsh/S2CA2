### Processing on the Monthly GoT data to remove false positives
### Output is a processed csv will import this to MongoDB

```python
#Read in month of gameofthrones data
GoT_df = pd.read_csv('C:\\data files\\May2016 GOT.csv')
```

```python
GoT_df.head()
```

```python
#Drop all where hashtag != gameofthrones

GoT_df['hashtags'] = GoT_df['hashtags'].str.lower()
GoT_df = GoT_df[GoT_df['hashtags'].str.contains('gameofthrones')]
```

```python
#Convert Year column to a datetime objest
GoT_df['tweet date'] = pd.to_datetime(GoT_df['tweet date'])
GoT_df['date'] = GoT_df['tweet date'].dt.date
```
